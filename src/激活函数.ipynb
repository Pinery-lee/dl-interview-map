```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    # Sigmoid 导数的数学特性：f'(x) = f(x) * (1 - f(x))
    s = sigmoid(x)
    return s * (1 - s)

# 1. 准备数据
x = np.linspace(-10, 10, 1000) # 在 -10 到 10 之间生成 1000 个点
y = sigmoid(x)
y_der = sigmoid_derivative(x)

# 2. 设置绘图风格
plt.figure(figsize=(10, 6))
plt.grid(True, linestyle='--', alpha=0.6)

# 3. 绘制 Sigmoid 曲线
plt.plot(x, y, label='Sigmoid', color='#1f77b4', linewidth=2)

# 4. 绘制导数曲线
plt.plot(x, y_der, label='Derivative (Gradient)', color='#ff7f0e', linewidth=2, linestyle='--')

# 5. 添加装饰
plt.title('Sigmoid Function and its Derivative', fontsize=14)
plt.xlabel('Input (x)', fontsize=12)
plt.ylabel('Output / Gradient', fontsize=12)
plt.axhline(0, color='black', linewidth=0.5) # x轴
plt.axvline(0, color='black', linewidth=0.5) # y轴
plt.legend(loc='upper left', fontsize=12)

# 打印导数的最大值，帮助理解梯度消失
print(f"Sigmoid 导数的最大值是: {np.max(y_der):.4f} (发生在 x=0 时)")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def tanh(x):
    return np.tanh(x)

def tanh_derivative(x):
    # Tanh 导数的数学特性：f'(x) = 1 - f(x)^2
    return 1 - np.power(tanh(x), 2)

# 1. 准备数据
x = np.linspace(-10, 10, 1000)
y = tanh(x)
y_der = tanh_derivative(x)

# 2. 绘图设置
plt.figure(figsize=(10, 6))
plt.grid(True, linestyle='--', alpha=0.6)

# 3. 绘制 Tanh 曲线 (输出范围在 -1 到 1 之间)
plt.plot(x, y, label='Tanh', color='#2ca02c', linewidth=2)

# 4. 绘制导数曲线
plt.plot(x, y_der, label='Derivative (Gradient)', color='#d62728', linewidth=2, linestyle='--')

# 5. 修饰
plt.title('Tanh Function and its Derivative', fontsize=14)
plt.xlabel('Input (x)', fontsize=12)
plt.ylabel('Output / Gradient', fontsize=12)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend(loc='upper left', fontsize=12)

# 打印关键数值
print(f"Tanh 导数的最大值是: {np.max(y_der):.4f} (发生在 x=0 时)")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    # ReLU 导数：x > 0 时为 1，x <= 0 时为 0
    # 注意：在数学上 x=0 处不可导，在工程实现中通常默认为 0
    return np.where(x > 0, 1, 0)

# 1. 准备数据
x = np.linspace(-10, 10, 1000)
y = relu(x)
y_der = relu_derivative(x)

# 2. 绘图设置
plt.figure(figsize=(10, 6))
plt.grid(True, linestyle='--', alpha=0.6)

# 3. 绘制 ReLU 曲线
plt.plot(x, y, label='ReLU', color='#9467bd', linewidth=2)

# 4. 绘制导数曲线
plt.step(x, y_der, label='Derivative (Gradient)', color='#e377c2', linewidth=2, linestyle='--')

# 5. 修饰
plt.title('ReLU Function and its Derivative', fontsize=14)
plt.xlabel('Input (x)', fontsize=12)
plt.ylabel('Output / Gradient', fontsize=12)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.ylim(-0.5, 5) # 为了更好地观察导数和输出
plt.legend(loc='upper left', fontsize=12)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def leaky_relu(x, alpha=0.05):
    # 当 x >= 0 时输出 x，当 x < 0 时输出 alpha * x
    return np.where(x >= 0, x, alpha * x)

def leaky_relu_derivative(x, alpha=0.05):
    # 当 x >= 0 时导数为 1，当 x < 0 时导数为 alpha
    return np.where(x >= 0, 1, alpha)

# 1. 准备数据
x = np.linspace(-10, 10, 1000)
alpha_val = 0.05
y = leaky_relu(x, alpha_val)
y_der = leaky_relu_derivative(x, alpha_val)

# 2. 绘图设置
plt.figure(figsize=(10, 6))
plt.grid(True, linestyle='--', alpha=0.6)

# 3. 绘制 Leaky ReLU 曲线
plt.plot(x, y, label=f'Leaky ReLU (alpha={alpha_val})', color='#d62728', linewidth=2)

# 4. 绘制导数曲线
plt.step(x, y_der, label='Derivative (Gradient)', color='#1f77b4', linewidth=2, linestyle='--')

# 5. 修饰
plt.title(f'Leaky ReLU Function and its Derivative (alpha={alpha_val})', fontsize=14)
plt.xlabel('Input (x)', fontsize=12)
plt.ylabel('Output / Gradient', fontsize=12)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.ylim(-0.5, 5)
plt.legend(loc='upper left', fontsize=12)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def silu(x):
    # SiLU(x) = x * sigmoid(x)
    return x * sigmoid(x)

def silu_derivative(x):
    # SiLU 的导数公式：f'(x) = f(x) + sigmoid(x) * (1 - f(x))
    # 也可以写成：sigmoid(x) * (1 + x * (1 - sigmoid(x)))
    s = sigmoid(x)
    return s * (1 + x * (1 - s))

# 1. 准备数据
x = np.linspace(-10, 10, 1000)
y = silu(x)
y_der = silu_derivative(x)

# 2. 绘图设置
plt.figure(figsize=(10, 6))
plt.grid(True, linestyle='--', alpha=0.6)

# 3. 绘制 SiLU 曲线
plt.plot(x, y, label='SiLU (Swish)', color='#17becf', linewidth=2)

# 4. 绘制导数曲线
plt.plot(x, y_der, label='Derivative (Gradient)', color='#bcbd22', linewidth=2, linestyle='--')

# 5. 修饰
plt.title('SiLU (Swish) Function and its Derivative', fontsize=14)
plt.xlabel('Input (x)', fontsize=12)
plt.ylabel('Output / Gradient', fontsize=12)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend(loc='upper left', fontsize=12)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def gelu(x):
    # GELU 的常用近似公式
    return 0.5 * x * (1 + np.tanh(np.sqrt(2 / np.pi) * (x + 0.044715 * np.power(x, 3))))

def gelu_derivative(x):
    # 这里使用数值微分来获取导数，展示其平滑特性
    dx = 1e-7
    return (gelu(x + dx) - gelu(x)) / dx

# 1. 准备数据
x = np.linspace(-5, 5, 1000)
y = gelu(x)
y_der = gelu_derivative(x)

# 2. 绘图设置
plt.figure(figsize=(10, 6))
plt.grid(True, linestyle='--', alpha=0.6)

# 3. 绘制 GELU 曲线
plt.plot(x, y, label='GELU', color='#1f77b4', linewidth=2)

# 4. 绘制导数曲线
plt.plot(x, y_der, label='Derivative (Gradient)', color='#ff7f0e', linewidth=2, linestyle='--')

# 5. 修饰
plt.title('GELU Function and its Derivative', fontsize=14)
plt.xlabel('Input (x)', fontsize=12)
plt.ylabel('Output / Gradient', fontsize=12)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend(loc='upper left', fontsize=12)

plt.show()
```
